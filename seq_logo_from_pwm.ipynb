```python
import numpy as np
from utils import datasets
import gzip
import csv
```

```python
# dataset
path = './data/encode/'
dataset_names = datasets(path)
```

```python
dataset = dataset_names[0][2]
```

```python
data = []

with gzip.open(dataset, 'rt') as f:
    next(f)
    reader = csv.reader(f, delimiter = '\t')
    for row in reader:
        # print(row[2])
        data.append(row[2])
```

```python
pfm = np.zeros((4, len(data[0]))) # A C G T -> 4 by len(data[0])
not_defined = 0
```

```python
for i in range(len(data)):
    for j in range(len(data[0])):
        if data[i][j] == 'A':
            pfm[0][j] = pfm[0][j] + 1
        elif data[i][j] == 'C':
            pfm[1][j] = pfm[1][j] + 1
        elif data[i][j] == 'G':
            pfm[2][j] = pfm[2][j] + 1
        elif data[i][j] == 'T':
            pfm[3][j] = pfm[3][j] + 1
        else:
            print("!")
            not_defined = not_defined + 1
```

```python
# error check

for i in range(len(data[0])):
    if(pfm[0][i] + pfm[1][i] + pfm[2][i] + pfm[3][i] != 5084):
        print("?")
```

```python
ppm = pfm/len(data)
```

```python
# zero value checking

for i in range(len(ppm)):
    for j in range(len(ppm[0])):
        if (ppm[i][j] == 0):
            print("?")
```

```python
for i in range(len(data[0])):
    if(round(ppm[0][i] + ppm[1][i] + ppm[2][i] + ppm[3][i], 10) != 1):
        print(i)
```

```python
information = np.zeros(len(ppm[0]))
```

```python
for j in range(len(information)):
    entropy_j = 0
    for i in range(4):
        entropy_j = entropy_j + ppm[i][j] * np.log2(ppm[i][j])
    information[j] = 2 + entropy_j
```

```python
seq_logo = np.zeros((4, len(data[0])))
```

```python
for i in range(len(seq_logo)):
    for j in range(len(seq_logo[0])):
        seq_logo[i][j] = ppm[i][j] * information[j]
```
